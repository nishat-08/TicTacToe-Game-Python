```python
# Function for print out the board and set up the board as list
# a 3 by 3 board representation
# use index position 1-9 corresponds with a number on a number board

from IPython.display import clear_output

def display_board(board):
    # to see the one version of board
    
    clear_output() 
    
    # Represent of board with index position
    
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3]) 
    print('   |   |')
    print ('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print ('-----------')
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    
# Function for player input
# To assign players markers as 'X' or 'O'
    
def player_input():
    
    '''OUTPUT = (Player 1 marker, Player 2 marker)'''
    
    marker = ' '
    
    # Used while loop so that we can continually ask until get a correct answer
    while marker != 'X' and marker != 'O':
        marker = input("Player1: Choose 'X' or 'O': ").upper()
        
    if marker == 'X':
            
        return ('X','O')
        
    elif marker == 'O':
            
        return ('O','X')

# Function for assigns marker to the board
# It takes board, marker and positon (1-9)

def place_marker(board, marker, position):
    
    board[position] = marker
    
# Function for win check
# It takes board and a marker 'X' or 'O' to check if that marker has won


def win_check(board,mark):
    
    # To win Tic Tac Toe we need to check 3 combination
    # Check Rows to see if they all share the same marker
    # Check Columns to see if marker matches
    # Check 2 diagonals to see if marker matches
    # It use board index position to check the marker matches
    
    return((board[1] == mark and board[2] == mark and board[3] == mark ) or # across the top
           (board[4] == mark and board[5] == mark and board[6] == mark) or  # across the middle
           (board[7] == mark and board[8] == mark and board[9] == mark) or  # across the bottom
           (board[1] == mark and board[4] == mark and board[7] == mark) or  # down the left side
           (board[2] == mark and board[5] == mark and board[8] == mark) or  # down the middle
           (board[3] == mark and board[6] == mark and board[9] == mark) or  # down the right side
           (board[1] == mark and board[5] == mark and board[9] == mark) or  # diagonal
           (board[3] == mark and board[5] == mark and board[7] == mark))    # diagonal

# Function to randomly decide which plyer goes first 
# Used random module randint() function [Random integer function of random module]

import random

def choose_first():
    
    flip = random.randint(0,1) # It is randomly going to choose either 0 or 1
    
    if flip == 0:
        return 'player 1'
    else:
        return 'player 2'

# Function to check if a space on the board is freely available
# It return a boolean
# Takes board and position

def space_check(board,position):
    
    return board[position] == ' '

# Function to check if the board is full 
# Its return a boolean
# Takes board and use function range(), space_check()

def full_board_check(board):
    
    for i in range(1,10): # 9 spaces on the board starting from 1 to upto 10 but not including 10
        if space_check(board,i):
            return False
    # Board is full is we return True
    return True

# Function for asking the next position from player
# Check if the space is free using space_check function


def player_choice(board):
    
    position = 0 # place holder nothing goes on index 0
    
    while position not in range(1,10) or not space_check(board,position):
        position = int(input('Choose a position (1-9): '))
    
    return position

# Function to ask the player if they want to play again or not
# Its return boolean

def replay():
    
    choice = input("Play agein? Enter Yes or No")
    
    return choice == 'Yes'

# Conditions and functions to run the game start here

# While loop to keep running the game

print('Welcome to Tic Tac Toe')

while True:
    
    # Play the game
    
    # Set Board, Whos first, choose marker 'X' or 'O'
    board = [' ']*10 # list of empty string
    player1_marker,player2_marker = player_input()
    
    turn = choose_first()
    print(turn + ' will go first')
    
    play_game = input ('Ready to Play? y or n?')
    
    if play_game == 'y':
        game_on = True
    else:
        game_on = False
    
    # Game Play
    
    while game_on:
        
        # Player 1 Turn
        if turn == 'Player 1':
            
            # Show thr board
            display_board(board)
            
            # Choose a position
            position = player_choice(board)
            
            # Place the marker on the position
            place_marker(board,player1_marker,position)
            
            # Check if they won
            if win_check(board, player1_marker):
                display_board(board)
                print('PLAYER 1 HAS WON!!')
                game_on = False
            
            # Check if there is a tie
            else:
                if full_board_check(board):
                    display_board(board)
                    print('TIE GAME!!')
                    game_on = False
                # No tie and no win? Then next player's turn
                else:
                    turn = 'Player 2'
            
        
        
        # Player 2 Turn
        else:
            # Show thr board
            display_board(board)
            
            # Choose a position
            position = player_choice(board)
            
            # Place the marker on the position
            place_marker(board,player2_marker,position)
            
            # Check if they won
            if win_check(board, player2_marker):
                display_board(board)
                print('PLAYER 2 HAS WON!!')
                game_on = False
            
            # Check if there is a tie
            else:
                if full_board_check(board):
                    display_board(board)
                    print('TIE GAME!!')
                    game_on = False
                else:
                    turn = 'Player 1'
            
            
      
        
    if not replay():
        break
# Break out the while loop on replay() function       
```
